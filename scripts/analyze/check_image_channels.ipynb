```python
COLORS = ['red', 'green', 'blue', 'yellow']
def load_rgby(image_dir, image_id, suffix='jpg', in_channels=4):
    images = [
        cv2.imread(f'{image_dir}/{image_id}_{color}.{suffix}', cv2.IMREAD_GRAYSCALE) for color in COLORS[0:in_channels]
    ]
    for image in images:
        if image is None:
            return None
    rgby = np.stack(images, axis=-1)
    return rgby

trang_img_ori = load_rgby(image_dir="/scratch/users/xikunz2/HPA_SingleCellClassification/kaggle-dataset/CAM_images/images/train", image_id="492_D3_2", suffix='png')
trang_img_ori.shape
```

```python
trang_img_ori.min(), trang_img_ori.max()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(10,6))
for i, ax in enumerate(axes):
    ax.imshow(trang_img_ori[:, :, i], cmap="gray")
```
