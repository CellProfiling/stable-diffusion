Check whether a protein always has the same location labels across different images

```python
from collections import defaultdict
import pickle
```

```python
with open("/data/wei/hpa-webdataset-all-composite/HPACombineDatasetInfo.pickle", "rb") as in_file:
    data_info = pickle.load(in_file)
len(data_info), data_info[0]
```

```python
prot2locs = defaultdict(set)
for x in data_info:
    prot = x["gene_names"]
    if str(prot) != "nan":
        prot2locs[prot].add(x["locations"])
len(prot2locs)
```

```python
for prot, locs in prot2locs.items():
    if len(locs) > 1:
        print(prot, locs)
```

The conclusion is that a protein can have different location labels across different images.
