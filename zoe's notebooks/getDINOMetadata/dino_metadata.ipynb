```python
import glob
import numpy as np
import pandas as pd
```

```python
lines = []
dir = "/scratch/groups/emmalu/JUMP_HPA_validation/2024-02-01T14-51-40_jump__ldm__vq4__ref45__o123__fov512_0.5/DINO_images/real/BR00116995/images/"
with open("dino_metadata_test.txt") as f:
    lines = f.readlines()
    lines = [dir+ line.strip() for line in lines]
    print(lines)
```

```python
cells = []
subtiles = []
wells = []
plates = []
wells = []
fovs = []
for path in lines:
    splits = path.split("/")[-1].split("_")
    plates.append(path.split("/real/")[1].split("/")[0])
    wells.append(splits[0].split("f")[0])
    fovs.append(splits[0].split("f")[1])
    subtiles.append(splits[1])
    cells.append(splits[-1].split(".png")[0])
        
```

```python
print(np.concatenate([[lines], [plates], [wells], [fovs], [subtiles], [cells]]).T.shape)
df = pd.DataFrame(np.concatenate([[lines], [plates], [wells], [fovs], [subtiles], [cells]]).T)
df.columns = ["Image_Name", "plates", "wells", "fovs", "subtiles", "cells"]
df["channel_width"] = 256
df["Target"] = np.nan
display(df)
```

```python
df.to_csv("dino_metadata_test.csv", index=False)
```
